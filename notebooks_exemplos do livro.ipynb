 O total de bits a serem usados na cache  a soma dos bits de dados (produto da quantidade de linhas pela largura em bits de cada uma)
mais os bits usados na coluna tag, cujo valor existe para cada linha.

Assim, temos:
### T (total de bits da cache) = Td (total de bits para a parte de dados) + Tt (total de bits dos tags)

#### Calcular o total de bits para a parte de dados (Td):
Td= capacidade total da cache em bits

A capacidade da cache  de 32KB, e cada byte tem 8 bits. Portanto, podemos calcular Td multiplicando a capacidade da cache (em bytes) em bits:



```python
#Td= capacidade total da cache em bits
capacidade_total= 32*(2**10)

#transforma o byte para bits multiplicando por 8
td = capacidade_total * 8

print("o total de bits nos dados : ", td)
```

##### Calcular o total de bits das tags (Tt):
O total de bits dos tags  calculado multiplicando a quantidade de linhas pela largura do campo tag:

Tt = Quantidade de linhas * Largura do campo tag

###### Calculo da quantidade de linhas na cache:
A quantidade de linhas na cache  determinada dividindo a capacidade da cache pela largura das linhas:

###### Calculo da largura do campo tag:
A largura do campo tag  determinada pela quantidade de blocos na memria principal (MP) 
dividida pela quantidade de linhas na cache.

Primeiro, precisamos calcular a quantidade de blocos na MP:

```python
import math

largura_das_linhas = 8
capacidade_da_cache = 32 * (2**10)
quantidade_de_linhas = capacidade_da_cache / largura_das_linhas

print("a quatidade de linhas : ", quantidade_de_linhas )

largura_celula= 8
cacapacidade_mp= 16*(2**20) 
Quantidade_de_blocos = cacapacidade_mp / largura_celula


blocos_por_linha= Quantidade_de_blocos / quantidade_de_linhas

print("a quantidade de tags ", blocos_por_linha)

largura_tag = math.log2(blocos_por_linha)

print("potanto a largura da tag  :", largura_tag )

Tt= quantidade_de_linhas * largura_tag
print("o total de bits das tags : ", Tt)

total
```
